```python
import fastf1
import fastf1.plotting
import seaborn as sns
from matplotlib import pyplot as plt

# The misc_mpl_mods option enables minor grid lines which clutter the plot
fastf1.plotting.setup_mpl(misc_mpl_mods=False)
```

```python
def lap_time_scatter_plot(driver, session):
    driver_laps = session.laps.pick_driver(driver).pick_quicklaps().reset_index()
    fig, ax = plt.subplots(figsize=(8, 8))

    sns.scatterplot(data=driver_laps,
                    x="LapNumber",
                    y="LapTime",
                    ax=ax,
                    hue="Compound",
                    palette=fastf1.plotting.COMPOUND_COLORS,
                    s=80,
                    linewidth=0,
                    legend='auto')

    ax.set_xlabel("Lap Number")
    ax.set_ylabel("Lap Time")

    # The y-axis increases from bottom to top by default
    # Since we are plotting time, it makes sense to invert the axis
    ax.invert_yaxis()
    plt.suptitle("NOR Laptimes in the 2023 Austria Grand Prix")

    # Turn on major grid lines
    plt.grid(color='w', which='major', axis='both')
    sns.despine(left=True, bottom=True)

    plt.tight_layout()
    ax.legend()
    plt.suptitle(f"Lap time scatter plot \n "f"{driver} {session.event['EventName']} {session.event.year} Race")
    plt.figure().set_size_inches(50000, 500000, forward=True)
    fig.set_dpi(1000)

    plt.show()
```

```python
lap_time_scatter_plot(drivers[1], session)
```
